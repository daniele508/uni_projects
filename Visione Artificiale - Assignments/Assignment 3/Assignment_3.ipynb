```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from skimage.metrics import peak_signal_noise_ratio as psnr
```

Definisco il metodo per il calcolo dell'errore MSE tra 2 immagini, cioè la somiglianza tra le 2 immagini.

Calcolo la differenza in valore assoluto tra i pixel delle 2 immagini per poi calcolarne la media dei valori al quadrato.

Il valore del MSE sarà la differernza quadratica media in valore assoluto dei pixel delle 2 immagini.

```python
def mse(img1, img2):
    diff=cv2.absdiff(img1, img2)
    mse=cv2.mean(diff**2)
    return mse[0]
```

Definisco la matrice aumentata "f_pad" dell'immagine tramite il padding.

Calcolo ED che è la differenza al quadrato tra l'immagine originale (image) e l'immagine filtrata (f).

Calcolo i pesi (c1,c2,c3,c4 ossia i valori dei quattro pixel vicini) per ogni pixel dell'immagine filtrata per definire la correttezza di una determinata soluzione. Il calcolo dei pesi consente di effettuare un filtraggio più preciso dell'immagine: i pixel più vicini a quello in esame hanno un peso maggiore rispetto ai pixel più lontani e questo può aiutare a conservare le caratteristiche importanti dell'immagine e a ridurre il rumore.

L'esponente "a" definisce quanto ciascun pixel influisce sul valore filtrato del pixel preso in considerazione.

Calcolo EP, ossia l'energia dei pixel vicini di f, come la somma dei pesi precedentemente calcolari, c1, c2, c3 e c4 per ciascun pixel, per arrivare al calcolo dell'energia totale E che è data dalla somma ED pesata con b ed EP pesata con c.

L'obiettivo dell'energia è quello di trovare la soluzione "f" che fornisce la miglior ricostruzione dell'immagine originale.

```python
def energy(image, f, a, b, c, f_pad):

    w,h=f_pad.shape

    ED=np.abs(image-f)**2
    c1=(np.abs(f-f_pad[0:w-2, 1:h-1]))**a
    c2=(np.abs(f-f_pad[2:w, 1:h-1]))**a
    c3=(np.abs(f-f_pad[1:w-1, 0:h-2]))**a
    c4=(np.abs(f-f_pad[1:w-1, 2:h]))**a
    EP=c1+c2+c3+c4
    E=ED*b+EP*c

    return E
```

Definisco l'algoritmo MRF.

L'algoritmo prende in input 'image'.

Definisco un numero massimo di iterazioni, i valori di a, b, e c e ricavo le dimensioni dell' immagine (w, h).
Creo una copia dell'immagine rumorosa e la chiamo f.

Creo una matrice a tre dimensioni (256,w,h) dove w e h sono le dimensioni dell'immagine di input. Questa matrice serve a contenere i valori di energia dell'immagine per ogni possibile valore di livello di grigio dei pixel, da 0 a 255.

L'algoritmo poi entra in un ciclo che itera fino a quando non viene raggiunto il numero massimo di iterazioni o fino a quando i valori di intensità dei pixel non cambiano più.

Utilizzo il metodo energy per calcolare l'energia dei pixel per assegnarla alla variabile EN. Quindi questa matrice 2D, che contiene il livello di energia per ogni pixel, è assegnata nell'l-esimo livello di grigio nella matrice E_n.

L'indice associato al livello di energia minimo lungo l'asse 0 corrisponde al livello di grigio del pixel. Quindi l'immagine f sarà confrontata con l'immagine risultante "ind_min" generata considerando il livello di grigio con il livello di energia minimo per ogni pixel, e se differiscono f verrà aggiornata e l'algoritmo procederà con la prossima iterazione, altrimenti l'algoritmo uscirà dal ciclo e terminerà.

```python
def MRF(image):
    max_iter=10
    w, h=image.shape
    changed=True

    a=0.9
    b=1
    c=b

    f=image.copy().astype(np.float32)
    E_n=np.zeros((256,w,h))

    for n in range(max_iter):
        changed=False
        f_pad=np.pad(f, (1,1))

        for l in range(256):
            EN=energy(image, l, a, b, c, f_pad)
            E_n[l]=EN

        ind_min=np.argmin(E_n, axis=0)

        if not np.array_equal(f, ind_min):
            f=ind_min
            changed=True
    
        if not changed:
            break

    return f.astype(np.uint8)
```

Applico il filtro MRF alle immagini con rumore.

Successivamente calcolo il Mean Square Error (MSE) a il Peak Signal-To-Noise Ratio (PSNR) tra l'immagine originale e l'immagine corrotta, e tra l'immagine corrotta e l'immagine filtrata, per ognuna della 10 immagini.

Infine gli errori vengono salvati in una matrice "errors" con dimensioni (10,4), dove ogni riga corrispone ad un immagine e ogni colonna ad ogni tipo di errore.

```python
i=0
errors=np.zeros((10,4))
for name in [9, 16, 28, 35, 52, 68, 80, 90, 96, 100]:
    ext=".bmp"
    image=cv2.imread("gr6/Noisy/R"+str(name)+ext)
    image=cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    f=MRF(image)
    cv2.imwrite("gr6/Filtered/MRF"+str(name)+".png", f)

    if name==80 or name==90 or name==35:
        ext=".jpg"

    org=cv2.imread("gr6/Original/"+str(name)+ext)
    org=cv2.cvtColor(org, cv2.COLOR_BGR2GRAY)

    errors[i,0] = mse(org, image)
    errors[i,1] = mse(f, image)
    errors[i,2]= psnr(org, image)
    errors[i,3]= psnr(f, image)
    i+=1
```

Stampo i valori per ogni tipo di errore

```python
for i in range(10):

    if i==0 or i==1 or 1==2 or i==3:
        filtro="gaussian"
    else:
        if i==6 or i==4 or i==5 or i==7:
            filtro="median"
        else:
            if i==8 or i==9:
                filtro="bilateral"

    print(f"computing errors for image {i+1} filtered by the {filtro} filter:")
    print(f"MSE (original image, corrupted image): {errors2[i,0]:.5f}")
    print(f"MSE (denoised image, corrupted image): {errors2[i,1]:.5f}")
    print(f"PSNR (original image, corrupted image): {errors2[i,2]:.5f}")
    print(f"PSNR (denoised image, corrupted image): {errors2[i,3]:.5f}\n")

    print(f"Computing errors for image {i+1} filtered using the MRF:")
    print(f"MSE (original image, corrupted image): {errors[i,0]:.5f}")
    print(f"MSE (denoised image, corrupted image): {errors[i,1]:.5f}")
    print(f"PSNR (original image, corrupted image): {errors[i,2]:.5f}")
    print(f"PSNR (denoised image, corrupted image): {errors[i,3]:.5f}\n")
```
